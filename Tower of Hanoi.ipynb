```python
#closed form
def towerClosed(m):
    ans = (2 ** m) - 1
    return ans

#recursive form
def towerRecursive(n):
    if n == 1:
        return 1
    else:     
        return (2 * towerRecursive(n-1)) + 1

#test loop
x = [towerClosed(i) for i in range(1,65)]
y = [towerRecursive(i1) for i1 in range(1,65)]
merged_list = [(x[i3],y[i3]) for i3 in range(0,len(x))]
print(merged_list)
```
